```python
#Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#1.DATA LOADING AND CLEANING
#Loading Dataset
sales = pd.read_csv("retail_sales.csv")
```

```python
sales.head()
```

```python
#Missing Values
sales.isnull().sum()
```

```python
sales = sales.dropna()
```

```python
sales.isnull().sum()
```

```python
#Remove Duplicates
sales = sales.drop_duplicates()
```

```python
sales.duplicated().sum()
```

```python
#2.TIME SERIES ANALYSIS
sales['Date'] = pd.to_datetime(sales['Date'])
```

```python
#3.Descriptive Statistics
sales.describe()
```

```python
sales['Sales'].mean()
```

```python
sales['Sales'].median()
```

```python
sales['Category'].mode()
```

```python
sales['Sales'].std()
```

```python
#Sales Ananlysis
sales['Month'] = sales['Date'].dt.month
```

```python
#VISUALIZATION
#Monthly Sales Trend
monthly_sales = sales.groupby('Month')['Sales'].sum()

monthly_sales.plot(kind='line')

plt.title('Monthly Sales Trend')

plt.xlabel('Month')

plt.ylabel('Sales')

plt.show()
```

```python
#Daily Sales Trend
daily_sales = sales.groupby('Date')['Sales'].sum()

daily_sales.plot()

plt.title('Daily Sales Trend')

plt.xlabel('Date')

plt.ylabel('Sales')

plt.show()
```

```python
sales['Category'].unique()
```

```python
import numpy as np

sales['Category'] = sales['Category'].replace(
    ['NaN?', 'Nan', 'Null'],
    np.nan
)
```

```python
sales = sales.dropna()
```

```python
sales.dropna(inplace=True)
```

```python
sales.isnull().sum()
```

```python
sales['Category'].unique()
```

```python
#Category Analysis
sales.groupby('Category')['Sales'].sum().plot(kind='bar')

plt.title('Sales by Category')

plt.xlabel('Category')

plt.ylabel('Sales')

plt.show()
```

```python
#Quantity Analaysis
sales.groupby('Category')['Quantity'].sum().plot(kind='bar')

plt.title('Quantity Sold by Category')

plt.xlabel('Category')

plt.ylabel('Quantity')

plt.show()
```

```python
#Profit Analysis
sales.groupby('Category')['Profit'].sum().plot(kind='bar')

plt.title('Profit by Category')

plt.xlabel('Category')

plt.ylabel('Profit')

plt.show()
```

```python
#Region Analysis
sales.groupby('Region')['Sales'].sum().plot(kind='bar')

plt.title('Sales by Region')

plt.xlabel('Region')

plt.ylabel('Sales')

plt.show()
```

```python
#Correlation Heatmap
corr = sales[['Sales','Quantity','Profit']].corr()

sns.heatmap(corr, annot=True)

plt.show()
```
